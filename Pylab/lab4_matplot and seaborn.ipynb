```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns #to manage data
```

```python
#command to load the iris dataset
iris = sns.load_dataset("iris")
iris.describe()
```

scatter plots are useful for visualizing relationships between two numerical variables.

```python
plt.figure(figsize =(10,6))
plt.scatter(iris["sepal_length"],iris["sepal_width"])
plt.title("Scatter plot of sepal features")
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.grid(True) #to add a grid

#I have to save before showing the plot
plt.savefig("scatterplot.png", dpi=900)
plt.show()
#often the plot are so small. We ca
```

histograms are useful for visualizing the distribution of a single numerical variable

```python
#working the same as scatterplots, but with more parameters 
plt.figure(figsize=(7,4))
plt.hist(iris['petal_length'], bins = 10,color = "red", edgecolor = "black")
#we want to add information of the axis 
plt.title("Petal length - Histogram")
plt.ylabel("frequency")
plt.xlabel("Petal length (cm)")
plt.grid(True)
plt.show()
```

```python
#working the same as scatterplots, but with more parameters 
plt.figure(figsize=(7,4))
plt.hist(iris['petal_length'], bins = 10,color = "skyblue", edgecolor = "black") #much better with soft color
plt.title("Petal length - Histogram")
plt.ylabel("frequency")
plt.xlabel("Petal length (cm)")
plt.grid(True)
plt.show()
```

boxplot are useful fot visualizing the distribution and outliers of a single numerical variable

```python
#something I'll use a lot in machine learning as well as the other courses 
plt.figure(figsize=(7,4))
plt.boxplot(iris['petal_length'])
#x label are not interesting in this case
plt.title("Petal length")
plt.ylabel("petal length (cm)")
plt.grid(True)
plt.show()

#minimum length is 1 and the maximum is 7: lower/upper buffle.
#the box rapresents the 95%, and the orange line is the centre of the normal distribution (it's not a normal distribution in this case)
```

line plots are useful for visualizing trends in data over time or over continuous variables

```python
plt.figure(figsize=(8,3.5))
plt.plot(iris['sepal_width'])
plt.title("sepal width lineplot")
plt.ylabel("sepal width (cm)")
 #plt.xaxis(False) #to remove the showing up of an axis 
plt.grid(True)
plt.show()
#we will have a set of points automatically connected through lines. 
#stock prices of continuous functions: this is the way to work 
#data scientists use more other kinds of plots 
```

subplots allow to disply multiple plots in a grid layout 

```python
#something more advanced now. 
#combining more then a plot 
fig,axs = plt.subplots(2,2,figsize = (14,10)) #returns a figure and an array
axs[0,0].scatter(iris["sepal_length"],iris["sepal_width"])
axs[0,0].set_title("Scatter plot of sepal features")

axs[0,1].hist(iris['petal_length'], bins = 10,color = "skyblue", edgecolor = "black") #much better with soft color
axs[0,1].set_title("Petal length - Histogram")

axs[1,0].hist(iris['petal_length'], bins = 10,color = "red", edgecolor = "black")
axs[1,0].set_title("Petal length - Histogram")

axs[1,1].plot(iris['sepal_width'])
axs[1,1].set_title("sepal width lineplot")

plt.tight_layout()
fig.savefig("subplots.png")
plt.show()
```

a pairplot displays pariwise relationships in a dataset

```python
#seaborn: to present results without spending time on the design of your plots 
sns.pairplot(iris, hue="species") #dataset as input, kinf of graphic I want to have 
plt.show()
#we have 14 plots to check. We have many information. 
#legend on the colors on the right 
#how to read the pairplot: on the diagonal I have the distribution for the specific features 
#all the other plots are relating different features: we became aware of clustering!!
```

a boxplot is used to display the distribution and outliers of a single or multiple numerical variables 

```python
#with seaborn we can make much better boxplot 
sns.boxplot(iris,x='species', y='petal_width' ) #data, what do we want on the x and y axis 
plt.show()
#diamonds rapresent outliars 
```

a violin plot combines a boxplot with a kernel density estimate to display the distribution of data 

```python
sns.violinplot(iris, x='species',y='petal_width')
#if we look above we saw the diamonds, while here there are the no outliars because they are included in the distribution. 
```

a swarmplot is a categorical plot with non-overlapping points 

```python
sns.swarmplot(iris, x='species',y='petal_width')
```

a count plot shows the counts of observations in each categorical bin using bars

```python
iris['sepal_length_cat'] = pd.cut(iris['sepal_length'],bins=3, labels = ["short",'medium', 'long'])
sns.countplot(data=iris,x='sepal_length_cat', hue="species")
```

a heatmap is used to display data values in a matrix format, with colors representing the magnitude of the values

```python
#most important for machine learing: CORRELATION PLOT
iris["setosa"].delete()
correlation_matrix = iris.corr()
#create heatmap with always annot set to True
sns.heatmap(correlation_matrix, annot=True,cmap = "coolwarm")

#diagonals: no worry. It takes two features and tells you how much they are correlated to the other features. Negative values: no correlation at all. 
#for high values (near 1) then the features are so correlated. the more one increases, the more the other increases

#it's fondamental to remove the noice of a dataset: remove the features that are on average uncorrelated to the others.
```

a joint plot displays a relationship between two variables along with the marginal distributions 

```python
#join plot 
sns.jointplot(data=iris, x='sepal_width', y='sepal_length', kind='scatter')
#it's a mess!
```

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', kind='scatter')
#it's not a mess because there is correlation in this case!!
#there is high correlation but there are two clusters too
```
