Leap Year Checker
Write a program that checks whether a given year is a leap year or not.

```python
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

Grade Calculator
Write a program that reads the marks of a student in three subjects and calculates the grade based on the average marks.

avg >= 90 then grade = A
80 <= avg < 90 then grade = B
70 <= avg < 80 then grade = C
...
up to F

```python
#my solution
a,b,c = int(input("enter the first mark")), int(input("enter the second mark")),int(input("enter the third mark"))
avg = (a+b+c)/3
if avg>=90:
    print("grade A")
elif 80<avg<90:
    print("grade B")
elif 70<avg<=80:
    print("grade C")
else:
    print("F")
```

```python
#official solution
marks1 = float(input("Enter marks for subject 1: "))
marks2 = float(input("Enter marks for subject 2: "))
marks3 = float(input("Enter marks for subject 3: "))

average = (marks1 + marks2 + marks3) / 3

if average >= 90:
    grade = 'A'
elif average >= 70:
    grade = 'B'
elif average >= 50:
    grade = 'C'
else:
    grade = 'F'

print(f"The grade is {grade}")
```

Pivot Table Creation
Given a DataFrame, create a pivot table to summarize the data. Group the data by 'Department', use 'Employee' as the index, and calculate the average salary for each department.

data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}

```python
#PIVOT: most complex stuff in pandas (not done in class)
import pandas as pd

# Given DataFrame
data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = df.pivot_table(values='Salary', index='Employee', columns='Department', aggfunc='mean', fill_value=0)
print(pivot_table)
```

Duplicate Data Handling
Given a DataFrame, find and drop duplicate rows.

```python
#duplicated: to take and remove the duplicates 
data = {'A': [1, 2, 2, 3], 'B': [4, 5, 5, 6]}
df = pd.DataFrame(data)
duplicated_rows = df[df.duplicated()]
print('Duplicated Rows:')
print(duplicated_rows)
df.drop_duplicates(inplace=True)
print('DataFrame after dropping duplicates:')
print(df)
```

Data Normalization
Given a DataFrame, normalize a column of data to have a mean of 0 and a standard deviation of 1.

```python
#data normalization 
#given a dataframe, normalize e column of data to have a mean of 0 and a sd of 1
data = {'A':[10,20,30,40,50]}
df = pd.DataFrame(data)
normalized_data = (df['A']-df['A'].mean())/df['A'].std()
print(normalized_data)
```
