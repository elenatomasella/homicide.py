```python
#sorted by an axis
iris_sorted = iris.sort_index(axis=1,ascending=True).head(10)
#axis must be equal to 0 or 1 and it defines the order: by row or by colomn respectively
#ascending to have the elements ordered from the smaller to the greater 
print(iris_sorted)
#we are switching the orders of the columns, maybe for visualizations purposes 
```

```python
#it's possible to switch the assigned order
iris_sorted = iris.sort_index(axis=1, ascending=False).head(10)
#in this case we have from the greater to the smaller 
print(iris_sorted)
```

```python
iris_sorted = iris.sort_index(axis=0,ascending=False).head(10)
#here we are sorting in terms of rows indeed the parameter axis is set to be 0 instead of 1
print(iris_sorted)
```

```python
iris_sorted = iris.sort_index(axis=0,ascending=True).head(10)
print(iris_sorted)
```

```python
#pandas also gives us a functions to get descriptive statistics for the dataframe 
summary_stats = iris.describe()

#display summary statistics 
print("Summary statistics:\n", summary_stats)
```

```python
#in this case we are counting the unique values in the column "Class": we are counting without repetitions 
unique_classes_count = iris['Class'].value_counts()

#display the count of unique species 
print("Unique Class Count:\n", unique_classes_count)
```

TAKE CARE OF WHAT'S BELOW
we can group the data as we prefer and print the funcion we are interested in directly.

```python
#we're going to make another: idea on the data set grouping values by something
grouped_data = iris.groupby('Class').mean()

##display the mean values for each species 
print("grouped Data with Mean: \n", grouped_data)
#this would be fundamental for machine learning 
```

```python
#play with the data frame
#it would be useful to have a portion of the dataset only 
# Select only the 'sepal_length' and 'sepal_width' columns
selected_columns = iris[['Sepal_Lenght', 'Sepal_Width']]

# Display the selected columns
print("Selected Columns:\n", selected_columns.head())
```

```python
#sort the dataset by sepal_length in descending order
sort_df =iris.sort_values(by = 'Sepal_Lenght', ascending=False)

#display the sorted data, descending order
print("Sorted Data by Sepal Length (Descending):\n", sort_df.head())

#we can also reverse the order using parametre ascending  = False
```

```python
#we can even order by multiple fields, for example by couple of columns 
sorted_df = iris.sort_values(by = ["Class", "Petal_Length"], ascending=[True, False])

#display the sorted data
print("Sorted Data by Class and Petal Length:\n", sorted_df)
```

```python
#MOST IMPORTANT: filtering the specific feature we want to study
filt_df = iris[iris["Petal_Width"]>1]

#display the filtered data
print("Filtered Data (Petal Length > 4.0):\n", filt_df.head())
```

```python
filtered_rows = iris[(iris["Petal_Length"]>1.5)&(iris["Petal_Width"]<3.0)]
#display the filtered rows
print("filtered Rows:\n", filtered_rows.head())
```

```python
#calculate the mean along axis 1
iris["Length mean"] = iris[["Sepal_Lenght", "Petal_Length"]].mean(axis=1)

iris.head()
```

```python
#create a cross tabulation between class and petal length categories 
cross_tab = pd.crosstab(iris['Class'], iris['Petal_Length'])

#display the cross tabulation
print("cross tabulation between class and petal length:\n", cross_tab)
```

```python
#create a new column "sepal_area" by multiplying "sepal_length" and "sepal_width"
iris["Sepal_Area"]=iris["Sepal_Lenght"]*iris["Sepal_Width"]
#display the dataframe with the new column 
print("dataframe with sepal area:\n", iris.head())
```

```python
#randomly sample 5 rows from the dataset
sampled_data = iris.sample(n=5, random_state=42)

#display the sampled data
print("Randomly sampled data:\n", sampled_data)
```

```python
#create a pivot table to show the mean "petal_length" for each "species"
#with pivot we completely change the dataset 
pivot_table = iris.pivot_table(index = "Class", values = "Petal_Length", aggfunc="mean") 

#index: we want to pivot the class
#aggregation function
#you can even mantain multiple values as a list 
pivot_table
```

```python
pivot_table = iris.pivot_table(index = "Class", values = ["Petal_Length", "Petal_Width"], aggfunc="mean")

pivot_table
```
