```python
# Calculate the sum of even numbers between two given numbers using a while loop
start = int(input("Enter the start number: "))
end = int(input("Enter the end number: "))
sum = 0
while start <= end:
    if start % 2 == 0:
        sum += start
    start += 1
print("Sum of even numbers:", sum)

#my solution
sum=0
a=int(input("your first number is: "))
b=int(input("your second number is: "))
while a<=b:
  if a%2==0:
    sum+=a
  a+=1
print(sum)
```
