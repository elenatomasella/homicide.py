## Find an element in a list

Write a function find_element(lst, v) that, given a list lst and a value v,
looks for v inside lst. 

If lst contains v, the function returns the index of its
first occurrence. Otherwise, it returns -1.

Example
- find_element([4, 10, 2, 10, 12], 10) shall return 1.
- find_element([4, 10, 2, 10, 12], 9) shall return - 1.

```python
#official solution 2
def find_element(lst, v):
    for index, element in enumerate(lst):
        if element == v:
            return index  # Return the index of the first occurrence of v
    return -1  # Return -1 if v is not found in lst

# Test the function with the provided examples
print(find_element([4, 10, 2, 10, 12], 10))  # Output: 1
print(find_element([4, 10, 2, 10, 12], 9))   # Output: -1
```

```python
#official solution 1
def find_element(input_list, goal_value):
  for index in range(len(input_list)):
    if input_list[index] == goal_value:
      return index
  return -1

print(find_element([4, 10, 2, 10, 12], 10))  # Output: 1
print(find_element([4, 10, 2, 10, 12], 9))   # Output: -1
```

## Reversed Words

Write a function reversed_words(s) that takes a string s as input, and returns a new string where each word in s is reversed, but the order of the words remains the same.

Example
- "hello world" => "olleh dlrow"
- "The quick brown fox" => "ehT kciuq nworb xof"

```python
def reversed_words(s):
    words = s.split(' ')
    reversed_words = [word[::-1] for word in words]
    result = ' '.join(reversed_words)
    return result
```

```python
# Test the function with the provided examples
s = "hello world"
result = reversed_words(s)
print(result)  # Output: "olleh dlrow"

s = "The quick brown fox"
result = reversed_words(s)
print(result)  # Output: "ehT kciuq nworb xof"
```

## Pad with Zeros

Write a function pad_with_zeroes(n, lst) whose parameters are an
integer n, and a list lst having at most n elements. 

The function shall
return a new list ret with length n, consisting in list lst "filled" with zeroes
on its head so that the number of elements is n. 

More precisely, the last
elements of ret shall be the elements of lst, in the same order. 

The first
"padding" elements of ret shall be zeroes. 

The input list lst shall not be
modified by the function.

Example
- a = [6, 5, 4, 3]
- b = pad_with_zeroes(10, a)
- print(b) # List with 10 elements: [0, 0, 0, 0, 0, 0, 6, 5, 4, 3]

```python
#PAD with zeros: my solution 
def pad_with_zeros(n,lst):
    for i in range(n-len(lst)):
        lst = [0] + lst
    return lst
```

```python
a= [6,5,4,3]
b = pad_with_zeros(10,a)
b
```

```python
#their solution
def pad_with_zeroes(n, lst):
    # Determine the number of zeros required
    padding_length = n - len(lst)
    # Create a list of zeros
    padding = [0] * padding_length
    # Concatenate the list of zeros with the input list
    ret = padding + lst
    return ret

# Test the function with the provided example
a = [6, 5, 4, 3]
b = pad_with_zeroes(10, a)
print(b)  # Output: [0, 0, 0, 0, 0, 0, 6, 5, 4, 3]
```

## Halve Even

Write a function halve_even_numbers(lst) that takes a list of integers lst
as a parameter. 

The function shall return a new list, containing the same
elements of input list lst, except for even numbers: even numbers shall
be halved. 

The function shall not modify the input list.

Example
- a = [10, 7, 22, 14, 9, 33]
- b = halve_even_numbers(a)
- print(b) # Expected output: [5, 7, 11, 7, 9, 33]

```python
#HALV EVEN: my solution 
def halve_even_number(lst):
    for i in range(len(lst)):
        if lst[i]%2==0:
            lst[i] = int(lst[i]/2) #you can use // to avoid the float
    return lst
```

```python
a = [10,7,22,14,9,33]
b = halve_even_number(a)
b
```

```python
#official solution
def halve_even_numbers(lst):
    # Create an empty list to hold the result
    result = []
    # Iterate through each element in the input list
    for element in lst:
        # Check if the element is even
        if element % 2 == 0:
            # If the element is even, halve it and append to the result list
            result.append(element // 2)
        else:
            # If the element is odd, keep it unchanged and append to the result list
            result.append(element)
    # Return the result list
    return result

# Test the function with the provided example
a = [10, 7, 22, 14, 9, 33]
b = halve_even_numbers(a)
print(b)  # Expected output: [5, 7, 11, 7, 9, 33]
```

## Find the intersection of two lists

Write a function intersection(l1, l2) that takes two lists as input, and
returns a new list, containing the common elements of l1 and l2.

Assume that all elements of l1 are distinct (i.e., there are no repetitions),
and that the same applies to l2.

```python
#FIND THE INTERSECTION OF TWO LISTS
def intersection(l1,l2):
    l3 = []
    for i in l1:
        if i in l2:
            l3.append(i)
    return l3
```

```python
l1 = [1,2,3,4]
l2 = [3,4,5]
intersection(l1,l2)
```

```python
#official solution
def intersection(l1, l2):
    # Create an empty list to hold the common elements
    common_elements = []
    # Iterate through each element in the first list
    for element in l1:
        # Check if the element exists in the second list
        if element in l2:
            # If it does, add it to the common elements list
            common_elements.append(element)
    # Return the common elements list
    return common_elements

# Test the function with an example
l1 = [1, 2, 3, 4, 5]
l2 = [4, 5, 6, 7, 8]
result = intersection(l1, l2)
print(result)  # Expected output: [4, 5]
```

## Max Locals

Write a function max_locals (lst) that receives a sequence list as
parameter and returns a list containing all the local maxima present in
sequence.

An element of a sequence is a local maximum if:
- it is not at one end of the list
- it is preceded and followed by strictly minor elements of the element
itself

```python
def max_locals (lst):
    local_maxima = []
    for i in range(1, len(lst)-1):
        if lst[i]>max(lst[i-1],lst[i+1]):
            local_maxima.append(lst[i])
    return local_maxima

# Test the function with an example
lst = [1, 3, 7, 1, 2, 6, 3, 2, 1]
result = max_locals(lst)
print(result)  # Expected output: [7, 6]
```

```python
#official solution
def max_locals(lst):
    # Create an empty list to hold the local maxima
    local_maxima = []
    # Iterate through the list, starting from the second element
    # and ending at the second-to-last element
    for i in range(1, len(lst) - 1):
        # Check if the current element is greater than both its
        # predecessor and its successor
        if lst[i] > lst[i - 1] and lst[i] > lst[i + 1]:
            # If it is, add it to the local maxima list
            local_maxima.append(lst[i])
    # Return the local maxima list
    return local_maxima

# Test the function with an example
lst = [1, 3, 7, 1, 2, 6, 3, 2, 1]
result = max_locals(lst)
print(result)  # Expected output: [7, 6]
```

## Bitonic

Write a function bitonic(lst) that receives a list as parameter and returns
True if the sequence is bitonic and False otherwise.

An array is said to be bitonic if the elements in the array are first strictly
increasing then strictly decreasing.

For example, the following sequences are bitonic:
- -3, 9, 11, 20, 17, 5, 1
- 3, 4, 1

```python
#my solution
def bitonic(lst):
    if len(lst)<3:
        return False
    else:
        i=0
        while i<len(lst) and lst[i]<lst[i+1]:
            i+=1
           
        while i>0 and i<len(lst)-1 and lst[i]<lst[i-1]:
            i+=1
        if i==len(lst):
            return True
        else:
            return False
# Test the function with the provided examples
lst1 = [-3, 9, 11, 20, 17, 5, 1]
lst2 = [3, 4,2]
print(bitonic(lst1))  # Expected output: True
print(bitonic(lst2))  # Expected output: True
```

```python
#my solution
def bitonic(lst):
    n = len(lst)
    # A sequence needs at least three elements to be bitonic
    if n < 3:
        return False
    # Find the peak element
    peak_found = False
    for i in range(1, n - 1):
        if lst[i] > lst[i - 1] and lst[i] > lst[i + 1]:
            peak_index = i
            peak_found = True
            break
    if not peak_found:
        return False  # No peak found, not bitonic
    # Check if elements before the peak are strictly increasing
    for i in range(1, peak_index):
        if lst[i] <= lst[i - 1]:
            return False
    # Check if elements after the peak are strictly decreasing
    for i in range(peak_index + 1, n - 1):
        if lst[i] <= lst[i + 1]:
            return False
    return True

# Test the function with the provided examples
lst1 = [-3, 9, 11, 20, 17, 5, 1]
lst2 = [3, 2, 9]
print(bitonic(lst1))  # Expected output: True
print(bitonic(lst2))  # Expected output: True
```

A palindromic string is a string that reads the same forwards and backwards. 

If there are multiple palindromic substrings of the same maximum length, return the one that appears first.

Example
"babad" = "bab" or "aba"
"cbbd" = "bb"
