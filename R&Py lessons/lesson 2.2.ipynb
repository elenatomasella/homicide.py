EXERCISE 1
Write a program to compute the Factorial.

```python
x = int(input('type the integer: '))
fact = 1
while x>1:
    fact = fact*x
    x-=1
print(fact)
```

#EXERCISE 2
Write a function to compute the Factorial, then call that function in the program to compute the factorial of a number given in input by the user.

```python
def factorial(x):
    fact = 1
    while x>1:
        fact = fact*x
        x-=1
    print(fact)
number = int(input("type your number: "))
factorial(number)
```

A lambda function is a small anonymous function. 
A lambda function can take any number of arguments, but can only have one expression.

A lambda function evaluates an expression for a given argument.
• The keyword lambda must come first.
• We give the function a value (argument) and then provide the operation (expression).
• A full colon (:) separates the argument and the expression.

```python
def function_name(x):
    return x+x
```

```python
lambda x: x+x
```

PROS
• Good for simple logical operations that are easy to understand.
• Good when we want to define a function that we will use just one time.
CONS
• They can only perform one expression.
• Bad for operations that would span more than one line in a normal def function.

EXERCISE 3
Return only the odd numbers from a list using a lambda function with filter(). 

```python
l = [5,4,3,7,8,5]
list(filter(lambda x: (x%2!=0),l))
```

```python
#print the first n odd numbers of the Fibonacci sequence
def Fib(n):
    if n==0:
        return "no sequence"
    elif n==1:
        return 0
    else:
        Fib_odd = []
        a,b = 0,1
        while len(Fib_odd)<n:
            if a%2!=0:
                Fib_odd.append(a)
            a,b = b,a+b
    for i in Fib_odd:
        print(i,end = " ")

Fib(9)
```
