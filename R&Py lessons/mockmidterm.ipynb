Define the calculator function that has as parameter two integers and a character (of the type +,-,*,/ in a string, so in inverted commas) and returns the result of the appropriate operation.

```python
def calculator(a,b,s):
    if s=="+":
        return a+b
    elif s=="-":
        return a-b
    elif s=="*":
        return a*b
    elif s=="/":
        return a/b
```

```python
calculator(5,6,"+")
```

The first two years of a dog are equal to 11.5 human years each. From the third are equal to 5 human years each. Write the function canine_years_count(integer) which takes an integer representing the number of canine years as parameter and prints out the equivalent in human years. (The result should always be a float).

```python
def canine_years_count(integer):
    if integer<=2:
        return f"the dog age is {integer *11.5}"
    else:
        return f"the dog age is {2*11.5+ (integer-2)*5}"
```

Write the function shipping_book that takes as input a floating point number (float) and two integers that represent respectively :
the cost of a book;
the number of copies purchased;
the shipping cost of the entire order, to be applied only if the number of copies purchased is less than 100.
The function calculates and prints the total cost of the purchase order.

```python
def shipping_book(c,n,s):
    if n<100:
        return c*n+s
    else:
        return c*n
```

Write a function vote that takes as parameter a vote (number) and translate it in a judgement (a string). Vote less than 18 is "Not sufficient", less than 22 is "Sufficient", less than 25 is "Good", less than 28 is "Great", 28 or more is "Excellent".

```python
def grade(number):
    if number<18:
        return "Not sufficient"
    elif number<22:
        return "Sufficient"
    elif number<25:
        return "Good"
    elif number<28:
        return "Great"
    else:
        return "Excellent"
```

```python
grade(25)
```

Write a Python program to count the number of vowels in a given string using a loop and a conditional statement.

```python
word = str(input("type your word"))
word = word.upper()
i,j=0,0
while i<len(word):
    if word[i] in ['A','E','I','O','U']:
        j+=1
    i +=1
print(f"there are {j} vowels")
```

Write a Python program to find the maximum number in a list without using the max() function. Utilise a loop and conditional statements.

```python
def maximum(lst):
    i,j=0,lst[0]
    while i<len(lst)-1:
        if lst[i]<lst[i+1]:
            j=lst[i+1]
        i+=1
    return j
```

```python
maximum([1,53,22,100,22])
```

Define the string_reverse function that has as parameter a string and returns the list with the characters of the string in reverse order (e.g. 'hello' becomes [o,l,l,e,h])

```python
def string_reverse(s):
    lst = []
    for i in s:
        lst.append(i)
    lst.reverse()
    return lst
```

```python
string_reverse('hello')
```

Write a Python program to check whether a given string is a palindrome or not. A palindrome is a word, phrase, number, or other sequences of characters that reads the same forward and backward (ignoring spaces, punctuation, and capitalisation).

```python
string = input("type your string: ")
string = string.lower()
chars, chars2 = [],[]
for i in range(0,len(string)):
  chars.append(string[i])
for i in range(0,len(string)):
  chars2.append(string[len(string)-1-i])
if chars2==chars:
  print("palindrome")
else:
  print("not palindrome")
```

Write a Python program to print all prime numbers within a given range. Make use of loops and conditionals to create an efficient algorithm.

```python
n = int(input("type a number"))
i,j=2,2
while i<=n:
    while j<=i and i%j!=0:
            j+=1
    if i==j:
          print(i, end=" ")
    j=2
    i+=1
```
