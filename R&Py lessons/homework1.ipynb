```python
#EXERCISE 5.1
#Given in input a natural number, say if it is prime or not.
n = int(input("type a natural number"))
i = 2
while (n%i!=0 and i<n):
        i+=1
if n-i==0:
    print("prime")
else:
    print("not prime")
```
