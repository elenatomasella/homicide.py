APIs  (application programming interfaces): type of software programming interface offering a service to other software components. 
A restful API is an architectural style for an application program interface (API) that uses HTTP request to access and use data.

HOW TO ACCESS A REST API
the request is made by an endpoint and a method. 
-the URL that delineates what data you are interacting with. it's tied to a specific resource within an API
-the method specifies how we are interacting with the resource located at the provided endpoint. Rest APIs can provide methods to CRUD (Create, Read, Upload, Delete)
-data: if you are using a method that involves changing data in a rest api, we will need to include a data payload with the request that inclued all data needed to create or modify exiting data
-headers: contain anu metadata that needs to be included with the request, such as authentication tokens, the content type that should be returned, and any caching policies
