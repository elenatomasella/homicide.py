```python
def factorial(x):
    fact = 1
    while x>1:
        fact = fact *x
        x-=1
    return fact

number = int(input())
factorial(number)
```

```python
#filter() with lambda()

#you can apply filter function to everthing: as second argument you must pass a list, and in the first the function to use as filter
a_list = [5,17,220,917,540]
output_list = list(filter(lambda x: (x%2 != 0), a_list))
print(output_list)
```
